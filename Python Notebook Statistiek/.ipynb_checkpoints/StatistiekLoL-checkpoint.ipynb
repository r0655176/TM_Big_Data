Dataset:

```python
# importing file
import pandas as pd
file = pd.read_excel('League.xlsx')
```

```python
# Showing columns
file
```

Statistiek:

```python
#Gemiddelde:
print("Averages:")
print("Duration: ",file['gameDuration'].mean())
print("Kills: ",file['kills'].mean())
print("Deaths: ",file['deaths'].mean())
print("Assists: ",file['assists'].mean())
print("Gold Earned: ",file['goldEarned'].mean())
print("Winrate: ",file['win'].mean())
```

```python
#Mediaan:
print("Medians:")
print("Duration: ",file['gameDuration'].median())
print("Kills: ",file['kills'].median())
print("Deaths: ",file['deaths'].median())
print("Assists: ",file['assists'].median())
print("Gold Earned: ",file['goldEarned'].median())
```

```python
#Modus:
print("Modus:")
print("Kills: ",file['kills'].mode())
print("Deaths: ",file['deaths'].mode())
print("Assists: ",file['assists'].mode())
```

```python
#Bereik:
print("Maximum / Minimun:")
print("Duration: ",file['gameDuration'].max(),"/",file['gameDuration'].min())
print("Kills: ",file['kills'].max(),"/",file['kills'].min())
print("Deaths: ",file['deaths'].max(),"/",file['deaths'].min())
print("Assists: ",file['assists'].max(),"/",file['assists'].min())
print("Gold Earned: ",file['goldEarned'].max(),"/",file['goldEarned'].min())
```

Standaarddeviatie:

```python
print("Standaarddeviatie")
print(file.std())
```

Correlation:

```python
print("Correlation:")
file.corr()
```

Graphics:

```python
file.sort_values(by="gameDuration", ascending=False).plot(x="gameDuration",y="goldEarned",title="Correlation duration/gold")
```

```python
file.sort_values(by="kills", ascending=False).plot.scatter(x="kills",y="goldEarned",title="Correlation kills/gold")
file.sort_values(by="assists", ascending=False).plot.scatter(x="assists",y="goldEarned",title="Correlation assists/gold")
```

Impact on data: Pyke (championID 555)

```python
pykegamesfile = file.query('championId==555')
pykegamesfile
```

```python
print("Duration: ",file['gameDuration'].mean()," vs ",pykegamesfile['gameDuration'].mean())
print("Kills: ",file['kills'].mean()," vs ",pykegamesfile['kills'].mean())
print("Deaths: ",file['deaths'].mean()," vs ",pykegamesfile['deaths'].mean())
print("Assists: ",file['assists'].mean()," vs ",pykegamesfile['assists'].mean())
print("Gold Earned: ",file['goldEarned'].mean()," vs ",pykegamesfile['goldEarned'].mean())
print("Winrate: ",file['win'].mean()," vs ",pykegamesfile['win'].mean())
```

```python
file.corr()
```

```python
pykegamesfile.corr()
```

```python
file.sort_values(by="kills", ascending=False).plot.scatter(x="kills",y="goldEarned",title="Correlation kills/gold")
file.sort_values(by="assists", ascending=False).plot.scatter(x="assists",y="goldEarned",title="Correlation assists/gold")
pykegamesfile.sort_values(by="kills", ascending=False).plot.scatter(x="kills",y="goldEarned",title="Correlation kills/gold Pyke")
pykegamesfile.sort_values(by="assists", ascending=False).plot.scatter(x="assists",y="goldEarned",title="Correlation assists/gold Pyke")
```

Conclusion:<br>
Games played on pyke tend to have more kills in exchange for assists.<br>
Longer games played on pyke have a better correlation with gold gained and kills earned.<br>
Games played on pyke have a higher focus on kills and assists to gain gold.

Impact on data: Win or Lose

```python
winfile = file.query('win==True')
winfile
```

```python
losefile = file.query('win==False')
losefile
```

```python
print("Duration: ",winfile['gameDuration'].mean()," vs ",losefile['gameDuration'].mean())
print("Kills: ",winfile['kills'].mean()," vs ",losefile['kills'].mean())
print("Deaths: ",winfile['deaths'].mean()," vs ",losefile['deaths'].mean())
print("Assists: ",winfile['assists'].mean()," vs ",losefile['assists'].mean())
print("Gold Earned: ",winfile['goldEarned'].mean()," vs ",losefile['goldEarned'].mean())
```

```python
winfile.corr()
```

```python
losefile.corr()
```

```python
winfile.sort_values(by="gameDuration", ascending=False).plot(x="gameDuration",y="goldEarned",title="Correlation duration/gold win")
losefile.sort_values(by="gameDuration", ascending=False).plot(x="gameDuration",y="goldEarned",title="Correlation duration/gold lose")
```

Conclusion:<br>
Wins have overall better scores even if they are on average shorter.<br>
Losses have an overall better correlation with assists earned than wins.<br>
Losses have a better focus on assists than wins to earn gold, but around the same focus on kills and deaths.
